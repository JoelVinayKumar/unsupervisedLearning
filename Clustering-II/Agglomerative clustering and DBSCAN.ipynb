**Basic concepts of Hierarchical Algorithms**

The hierarchical clustering means we can start from either singleton clusters, iteratively merge them into higher level clusters or we can start from one big macro cluster, iteratively split the bigger cluster into smaller ones, which will help us to generate a clustering hierarchy called as dendogram. 

And unlike K-Means we don't have to specify the number of clusters upfront. And this approach is divided into two types.

    1. Agglomerative clustering
    2. Divise Clustering

#### Agglomerative clustering

Agglomerative clustering refers to an algorithm that starts by declaring each point as its own cluster, and then merges the two most similar clusters iteratively until some stopping criterion is satisfied.

###### (AG)GLOMERATIVE (NES)TING:
Is an algorithm that uses a single-link method or more often referred to as nearest neighbor method. It keeps a dissimilarity matrix and continuously merges nodes that have least dissimilarity in other sense most nearest neighbors. And eventually all the similar nodes are merged into one cluster.

There are several linkage criteria that specify how the similar single-ton sets are measured. 
    
    1.Single Link
    2.Complete Link (diameter)
    3.Average Link
    4.Central Link (distance between cluster centroids)
    
<img src = "images/agnes.png">

In the beginning, each point is its treated as a single cluster. Then, in each step, the two clusters that are closest are merged. In the first four steps, two single point clusters are picked and these are joined into two-point clusters. In step four, one of the two-point clusters is extended to a third point, and so on. In step 9, there are only three clusters are remaining. Here 3 is our threshould and the algorithm stops. 

By the way the algorithm works AGNES cannot predict results for a newer datapoints. 

```python
# Applying AGGLOMERATIVE CLUSTERING on synthetic data
from sklearn.cluster import AgglomerativeClustering
from sklearn.datasets import make_blobs
X, y = make_blobs(random_state=1)
agg = AgglomerativeClustering(n_clusters=3)
assignment = agg.fit_predict(X)
plt.scatter(X[:, 0], X[:, 1], c=assignment, s=60)
```

As expected, the algorithm recovers the clustering perfectly. While the agglomerative clustering requires you to specify a number of clusters you want the algorithm to find, we will now see how agglomerative clustering methods helps with choosing the right number.

###### Hierarchical Clustering and Dendrograms

Agglomerative clustering being a type of hierarchical clustering helpful to look at all possible clusterings jointly. 

<img src="images/aglo.png">

Above image provides a detailed view of every possible cluster, but there is another approach called as Dendogram that can be used to visualize even an higher dimensional data.

```python
# import the dendrogram function and the ward clustering function from scipy
from scipy.cluster.hierarchy import dendrogram, ward
X, y = make_blobs(random_state=0, n_samples=15)
# apply the ward clustering to the data array X
# The scipy ward function returns an array that specifies the distances bridged when performing agglomerative
linkage_array = ward(X)
# now we plot the dendrogram for the linkage_array containing the distances between clusters
dendrogram(linkage_array);
# mark the cuts in the tree that signify two or three clusters
ax = plt.gca()
bounds = ax.get_xbound()
ax.plot(bounds, [9.5, 9.5], '--', c='k')
ax.plot(bounds, [4, 4], '--', c='k')
ax.text(bounds[1], 9.5, ' two clusters', verticalalignment='center', fontdict={'size': 15})
ax.text(bounds[1], 4, ' three clusters', verticalalignment='center', fontdict={'size': 15})
plt.title("Dendrogram")
```

Dendogram is similar to a tree, in this the leaf nodes are the data points and as we go to top we create parents by joining two nodes. And while interpretting a dendogram, the length of the branch also matters, the longer the length of the branch the more dissimilar the clusters are, which also mean we are clustering some very distant points. 

But still we cannot say that agglomerative clustering is the most superior algorithm. Even with AGNES also its difficult to cluster asymmetric data.

###### DBSCAN (Densitybased spatial clustering of applications with noise)

DBSCAN algorithm doesn't require the user to specify the number of clusters upfront. Since it is a density based clustering algorithm, DBSCAN can capture clusters of complex and asymmetric shapes. Additionally it can also identify the data points that won't be part of any cluster (Outliers).

The way DBSCAN works is by identifying the data points that are in **crowded** regions of the feature space, where many data points are close together. These regions are referred to as dense regions in feature space. The idea behind DBSCAN is ti identify clusters form dense regions of data, separated by regions that are relatively sparse.

Points that are within a dense region are called **core samples**.

The algorithm is defined by two parameters,

**Epsom** or EPS, which defines the maximum radius of the neighborhood, which says, if we have a point 'p' and we specifid the radius(EPS) as 1 cm. Then our aim would be to find out the points that are in the perimeter of 'p'.

**minimum points** is another parameter is used to define if a data point could be called as a core sample ot not. For a given point if you minimum number of points in the distance of eps then that point is called as a core sample. 

If a data point is called as core sample then it is given a cluster label of its own. The neighbors of the core sample are visited in turn to increase the cluster size until there are no more core-samples within the distance of EPS. Then a new data point is picked and similar procedure is followed.

By the end we will have three types of points,

**Core points**: A core sample, which has dense neighborhood.

**Border points** : Is within a cluster but cannot be treated as a core sample as it has comparatively less dense neighborhood.

**Noise / Outlier** : Not in a cluster.


Increasing EPS value means that more points will be included in a cluster. This makes clusters grow, but might also lead to multiple clusters joining into one. Increasing minimum samples means that fewer points will be core points, and more points will be labeled as outliers.

Setting EPS to be very small will mean that no points are core samples, and may lead to all points being labeled as noise. Setting EPS to be very large will result in all points forming a single cluster.

Finding a good setting for EPS is sometimes easier after scaling the data using Scaling as it will ensure that all features have similar ranges.

```python
# Applying DBSCAN for moon data 
from sklearn.datasets import make_moons
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import DBSCAN
import matplotlib.pyplot as plt

X, y = make_moons(n_samples=200, noise=0.05, random_state=0)

# scaling to centre the data
scaler = StandardScaler()
scaler.fit(X)
X_scaled = scaler.transform(X)

#Applying DBSCAN
model = DBSCAN()
clusters = model.fit_predict(X_scaled)
plt.scatter(X_scaled[:, 0], X_scaled[:, 1], c = clusters, s = 60)
```

```python
# Applying DBSCAN on faces dataset
import numpy as np
from sklearn.decomposition import PCA
from sklearn.datasets import fetch_lfw_people

people = fetch_lfw_people(min_faces_per_person=20, resize=0.7)
image_shape = people.images[0].shape

mask = np.zeros(people.target.shape, dtype=np.bool)
for target in np.unique(people.target):
    mask[np.where(people.target == target)[0][:50]] = 1

X_people = people.data[mask]
y_people = people.target[mask]

# scale the grey-scale values to be between 0 and 1
# instead of 0 and 255 for better numeric stability:
X_people = X_people / 255
print("Dataset shape: ", X_people.shape)

pca = PCA(n_components=100, whiten=True)
pca.fit_transform(X_people)
X_pca = pca.transform(X_people)
```

```python
# apply DBSCAN with default parameters
dbscan = DBSCAN()
labels = dbscan.fit_predict(X_pca)
print("Algo Op:", np.unique(labels))
```

We can see that all labels are -1, so all of the data was labeled as “noise”. We can try varying two things to improve the results, we can make EPS higher, to expand the neighborhood of each point, and decrease minimum samples, to consider smaller groups of points as clusters.

```python
dbscan = DBSCAN(min_samples=3)
labels = dbscan.fit_predict(X_pca)
print("Algo Op:", np.unique(labels))
```

Even by considering minimum samples as 3 also has termed the entire data as noise, we can now try varying EPS value

```python
dbscan = DBSCAN(min_samples=3, eps=15)
labels = dbscan.fit_predict(X_pca)
print("Algo Op:", np.unique(labels))
```

Specifying EPS values as 15 we got only a single cluster and noisy points. We can use this result and find out what the “noise” looks like compared to the rest of the data.

```python
# count number of points in all clusters and noise.
# bincount doesn't allow negative numbers, so we need to add 1.
# the first number in the result corresponds to noise points
np.bincount(labels + 1)
```

```python
noise = X_people[labels==-1]
fig, axes = plt.subplots(3, 9, subplot_kw={'xticks': (), 'yticks': ()}, figsize=(12, 4))
for image, ax in zip(noise, axes.ravel()):
    ax.imshow(image.reshape(image_shape), vmin=0, vmax=1)
```

Looking at these images we can guess why they were labeled as noise: 

- the image in the sixth image in the first row one has a person drinking from a glass, 
- there are images with hats, and the second to last image has a hand in front of the face. 
- The other images contain odd angles or crops that are too close or too wide.

```python
for eps in [1, 3, 5, 7, 9, 11, 13]:
    print("\neps=%d" % eps)
    dbscan = DBSCAN(eps=eps, min_samples=3)
    labels = dbscan.fit_predict(X_pca)
    print("Number of clusters: %s" % np.unique(labels))
    print("Clusters: %s" % np.bincount(labels + 1))
```

Interesting thing to note is that there is never more than one large cluster. There is at most one large cluster containing most of the points, and some smaller clusters. This indicates that there are not two or three different kinds of face images in the data that are very distinct, but that all images are more or less equally similar from the rest.
