###### Clustering
Clustering is the task of partitioning the dataset into groups, called clusters. The goal is to split up the data in such a way that points within a single cluster are very similar and points in different clusters are different. In much simpler sense Clustering is used to find groups of similar observations within a set of unlabeled data. 

###### K-Means
In this chapter, we’re going to study one type of clustering algorithm called k-means. It’s called kmeans because it finds k unique clusters, and the center of each cluster is the mean of the values in that cluster.

Before we get into k-means, let’s look at what cluster identification is. Cluster identification tells an algorithm, “Here’s some data. Now group similar things together and tell me about those groups.” The key difference from the supervised approach classification is that in classification you know what you’re looking for. That’s not the case in clustering. Clustering is sometimes called unsupervised classification because it produces the same result as classification but without having predefined classes.

With cluster analysis we will be trying to put similar things in a cluster and dissimilar things in a different clusters. This notion of similarity depends on a similarity measure used. Based on the use case, mean/median/mode will be used to cluster a given data.


**k-means** is an algorithm that will find k-clusters for a given dataset. The number of clusters k should be defined by user. Each cluster is described by a single point known as the **centroid**, means it’s at the center of all the points in a cluster. Typically if you have k-clusters then you will be having k-centroids respective to each cluster.

**pseudo code**
    
    1.Define the initial groups' centroids. This step can be done using different strategies. A very common one is to assign random values for the centroids of all groups. Another approach is to use the values of K different entities as being the centroids.
    
    2.Assign each entity to the cluster that has the "closest centroid". In order to find the cluster with the most similar centroid, the algorithm must calculate the distance between all the data points and each centroid.
    
    3.Recalculate the values of the centroids. The values of the centroid's fields are updated, taken as the average of the values of the data points' attributes that are part of the cluster.
    
    4.Repeat steps 2 and 3 iteratively until data points can no longer change groups.
    
the term "closest centroid" is mentioned, this implies some sort of distance measure should be used to determine the closest centroid of a data point. Most generally used metric is Euclidean distance. Instead we can also use Manhattan or any other metric. 

Similar to most of the machine learning algorithms, the optimal values of K-means' parameters found by minimizing its cost function. which is 
$$J = \sum_{k=1}^{K}\sum_{i=C_k}\parallel x_i - \mu_k\parallel^2$$

In the preceeding equation μ_k is the centroid of the k_th cluster. The cost function sums the distortions of the clusters. Each cluster's distortion is equal to the sum of the squared distances between its centroid and its  constituent instances. The distortion is small for compact clusters and large for clusters that contain scattered instances.

The parameters that minimize the cost function are learned as per the above pseudo code. 

```python
# Example
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans

#Generating synthetic data
X, y = make_blobs(random_state=1)
kmeans = KMeans(n_clusters=3)
kmeans.fit(X)
```

```python
# Now trying to predict labels
print(kmeans.predict(X))
```

```python
# If we plot that data
import matplotlib.pyplot as plt
plt.scatter(X[:, 0], X[:, 1], c=kmeans.labels_, s=60)
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], marker='^', s=100, linewidth=2, c = 'b')
plt.show()
```

```python
fig, axes = plt.subplots(1, 2)
# using two cluster centers:
kmeans = KMeans(n_clusters=2)
kmeans.fit(X)
assignments = kmeans.labels_
axes[0].scatter(X[:, 0], X[:, 1], c=assignments, s=60)

# using five cluster centers:
kmeans = KMeans(n_clusters=5)
kmeans.fit(X)
assignments = kmeans.labels_
axes[1].scatter(X[:, 0], X[:, 1], c=assignments, s=60);
plt.show()
```

###### Local optima
The cluster centroids are initialized randomly in K-Means. Due to which K-Means might not always recover the same resutlts.

Each cluster is solely defined by its centroid. Now if their initialization is altered they is possiblity of obtaining different clusters in every run and will derive random results, some times even not letting the algorithm not to converge and ditching it in Local optima. 

In cases where cluster centroids are initialized as per below image,
<img src = "images/local1.png"> 

K-Means will converge to local optimum as in below image yielding such irrelavant results. 
<img src="images/local2.png">

These clusters may be informative, but it is more likely that the top and bottom groups of observations are more informative clusters. To avoid local optima, K-Means is often repeated dozens or even hundreds of times. In each iteration, it is randomly initialized to different starting cluster positions.

Another important thing about K-Means is that the clusters derived by K-Means are convex shaped. As a results K-Means can only calculate simple shapes. k-Means also assumes that all clusters have the same “diameter” in some sense, it always draws the boundary between clusters to be exactly in the middle between the cluster centers. 


```python
X, y = make_blobs(random_state=0)
plt.scatter(X[:, 0], X[:, 1])
plt.show()
```

K-Means assumes that all directions are equally important for each cluster. In case of below example even though we can see that there are 3 clearly separable parts of the data but are stretched accross diagonal and are not symmetric in shape, K-Means only considers the distance to the nearest cluster and hence fails to handle such data.

```python
import numpy as np
# generate some random cluster data
X, y = make_blobs(random_state=170, n_samples=600)
rng = np.random.RandomState(74)
# transform the data to be streched
transformation = rng.normal(size=(2, 2))
X = np.dot(X, transformation)
# cluster the data into three clusters
kmeans = KMeans(n_clusters=3)
kmeans.fit(X)
y_pred = kmeans.predict(X)
# plot the cluster assignments and cluster centers
plt.scatter(X[:, 0], X[:, 1], c=y_pred)
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], marker='^', c=['b', 'r', 'g'], 
            s=60, linewidth=2)
plt.show()
```

```python
# Another example of a non-symmetric data
# generate synthetic two_moons data (with less noise this time)
from sklearn.datasets import make_moons
X, y = make_moons(n_samples=200, noise=0.05, random_state=0)
# cluster the data into two clusters
kmeans = KMeans(n_clusters=2)
kmeans.fit(X)
y_pred = kmeans.predict(X)
# plot the cluster assignments and cluster centers
plt.scatter(X[:, 0], X[:, 1], c=y_pred, s=60)
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], marker='^', c=['b', 'g'], 
            s=60, linewidth=2);
```

###### Evaluation
In majority cases K-value is specified by problem context.

Example:

A company that manufactures shoes might know that it is able to support manufacturing three new models. To understand what groups of customers to target with each model, it surveys customers and creates three clusters from the results. That is, the value of K was specified by the problem's context. Other problems may not require a specific number of clusters, and the optimal number of clusters may be ambiguous and in such scenarios the famous **elbow method** is used to figure out the optimal K-value.

We have another performance measure for clustering called as **Silhoutte coeffecient**, The silhouette coefficient is a measure of the compactness and separation of the clusters. It increases as the quality of the clusters increase. it is large for compact clusters that are far from each other and small for large, overlapping clusters. The silhouette coefficient is calculated per data point for a set of instances, it is calculated as the mean of the individual samples' scores. Silhoutte coeffecient per data points is calculated as follows.
    $$s = \frac{ba}{max(b,a)}$$
    
    a is the mean distance between the data points in the cluster. 
    b is the mean distance between the data points and the data points in the next closest cluster.

##### Image Quantization
Image quantization is a lossy compression method that replaces a range of similar colors in an image with a single color. Say we have differen shades in a single colour, such as, magenta, redwood, scarlet, ruby, carmine etc., then all those nearest shades will be replaced by single colour. 

Quantization reduces the size of the image file since fewer bits are required to represent the colors.


**In the following example, we will use clustering to discover a compressed palette for an image that contains its most important colors. We will then rebuild the image using the compressed palette.**

```python
try:
    import mahotas as mh
except:
    !conda install -c conda-forge mahotas
    import mahotas as mh
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.utils import shuffle

# reading image
actual_img = np.array(mh.imread("images/macau.jpg"), dtype=np.float64)/255
plt.title('Original Image')
plt.imshow(actual_img)
```

```python
# flattening the image
actual_shape = actual_img.shape
print("Shape of Image: {}".format(actual_shape))
width, height, depth = actual_img.shape
image_flattened = np.reshape(actual_img, (width * height, depth))
print("Shape of falttened image: {}".format(image_flattened.shape))

# We will now use K-Means to create 64 clusters from a sample of 1,000 randomly selected colors. Each of the 
# clusters will be a color in the compressed palette.

image_array_sample = shuffle(image_flattened, random_state=0)[:1000]
estimator = KMeans(n_clusters=64, random_state=0)
history = estimator.fit(image_array_sample)
```

```python
# now predict the cluster assignment for each pixel in the original image
cluster_assignments = estimator.predict(image_flattened)
```

```python
# recreate the compressed image from the compressed palatte and cluster assignments.
compressed_palette = estimator.cluster_centers_
compressed_img = np.zeros((width, height, compressed_palette.shape[1]))
label_idx = 0
for i in range(width):
    for j in range(height):
        compressed_img[i][j] = compressed_palette[cluster_assignments[label_idx]]
        label_idx += 1
        
plt.subplot(122)
plt.title('Original Image')
plt.imshow(actual_img)
plt.axis('off')
plt.subplot(121)
plt.title('Compressed Image')
plt.imshow(compressed_img)
plt.axis('off')
plt.show()
```

```python
import scipy.misc
scipy.misc.imsave('images/compressed_macau.jpg', compressed_img)
```

###### Clustering to learn features

Here we are will be combining both clustering and classification techniques to build a semi-supervised solution. We will use clustering to derive features from an unlabeled data and then use those newly learned features to build supervised classifier model.

Here we will be building an algorithm that does the task of classifying the images of doga and cats. we will cluster the descriptors extracted from all of the images to learn features. We will then represent an image with a vector with one element for each cluster. Each element will encode the number of descriptors extracted from the image that were assigned to the cluster. This approach is sometimes called the bag-of-features representation, as the collection of clusters is analogous to the bag-of-words representation seen for a text vocabulary.


We will use 1,000 images of cats and 1,000 images of dogs from the training set for Kaggle's Dogs vs. Cats competition. The dataset can be downloaded from https://www.kaggle.com/c/dogs-vs-cats/data. We will label cats as the positive class and dogs as the negative class.

Note that the images have different dimensions; since our feature vectors do not represent pixels, we do not need to resize the images to have the same dimensions. We will train using the first 60 percent of the images, and test on the remaining 40 percent

```python
# importing dependencies
import numpy as np
import mahotas as mh
from mahotas.features import surf
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import *
from sklearn.cluster import MiniBatchKMeans
import glob
```

First, we load the images, convert them to grayscale, and extract the SURF descriptors. SURF descriptors can be extracted more quickly than many similar features, but extracting descriptors from 2,000 images is still computationally expensive

```python
all_instance_filenames = []
all_instance_targets = []
for f in glob.glob('train/*.jpg'):
    target = 1 if 'cat' in f else 0
    all_instance_filenames.append(f)
    all_instance_targets.append(target)
surf_features = []
counter = 0
for f in all_instance_filenames:
#     print ('Reading image:', f)
    image = mh.imread(f, as_grey=True)
    surf_features.append(surf.surf(image)[:, 5:])
train_len = int(len(all_instance_filenames) * .60)
X_train_surf_features = np.concatenate(surf_features[:train_len])
X_test_surf_feautres = np.concatenate(surf_features[train_len:])
y_train = all_instance_targets[:train_len]
y_test = all_instance_targets[train_len:]
```

```python
n_clusters = 300
print ('Clustering', len(X_train_surf_features), 'features')
estimator = MiniBatchKMeans(n_clusters=n_clusters)
estimator.fit_transform(X_train_surf_features)
```

Next, we construct feature vectors for the training and testing data. We find the cluster associated with each of the extracted SURF descriptors, and count them using NumPy's binCount() function. The following code produces a 300-dimensional feature vector for each instance.

```python
X_train = []
for instance in surf_features[:train_len]:
    clusters = estimator.predict(instance)
    features = np.bincount(clusters)
    if len(features) < n_clusters:
        features = np.append(features, np.zeros((1, n_clusterslen(features))))
    X_train.append(features)
X_test = []
for instance in surf_features[train_len:]:
    clusters = estimator.predict(instance)
    features = np.bincount(clusters)
    if len(features) < n_clusters:
        features = np.append(features, np.zeros((1, n_clusterslen(features))))
X_test.append(features)
```

```python
# Finally use logistic regression model on the newer data and find out various metrics
clf = LogisticRegression(C=0.001, penalty='l2')
clf.fit(X_train, y_train)
predictions = clf.predict(X_test)
print (classification_report(y_test, predictions))
print ('Precision: ', precision_score(y_test, predictions))
print ('Recall: ', recall_score(y_test, predictions))
print ('Accuracy: ', accuracy_score(y_test, predictions))
```
