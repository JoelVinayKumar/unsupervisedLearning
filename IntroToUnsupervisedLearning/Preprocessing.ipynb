```python
# importing dependencies
import numpy as np
np.set_printoptions(suppress=True, precision=2)
from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split
```

```python
# loading cancer dataset
cancer = load_breast_cancer()
X_train, X_test, y_train, y_test = train_test_split(cancer.data, cancer.target,
random_state=1)
print(X_train.shape)
print(X_test.shape)
```

```python
# Applying Min-Max Scalar
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
scaler.fit(X_train)
```

```python
X_train_scaled = scaler.transform(X_train)
print("transformed shape: %s" % (X_train_scaled.shape,))
print("per-feature minimum before scaling:\n %s" % X_train.min(axis=0))
print("per-feature maximum before scaling:\n %s" % X_train.max(axis=0))
print("per-feature minimum after scaling:\n %s" % X_train_scaled.min(axis=0))
print("per-feature maximum after scaling:\n %s" % X_train_scaled.max(axis=0))
```

Data simply shifted and scaled and brought between zero and one.

```python
# scaling test data too.
X_test_scaled = scaler.transform(X_test)
```

```python
import matplotlib.pyplot as plt
# trying to visualize raw data
fig, axes = plt.subplots(1, 2, figsize=(13, 4))
axes[0].scatter(X_train[:, 0], X_train[:, 1], c='b', label="training set", s=60)
axes[0].scatter(X_test[:, 0], X_test[:, 1], marker='^', c='r', label="test set", s=60)
axes[0].legend(loc='upper left')
axes[0].set_title("original data")

# visualizing scaled data
axes[1].scatter(X_train_scaled[:, 0], X_train_scaled[:, 1], c='b', label="training set", s=60)
axes[1].scatter(X_test_scaled[:, 0], X_test_scaled[:, 1], marker='^', c='r', label="test set", s=60)
axes[1].set_title("scaled data")
plt.show()
```

First Panel has unscaled data and second panel has data scaled properly with MinMax Scalar. Data points are all shifted properly between 0 and 1. Minimum and Maximum values for the test data are not 0 and 1. 

```python
# what if data is not scaled properly
test_scaler = MinMaxScaler()
test_scaler.fit(X_test)
X_test_scaled_badly = test_scaler.transform(X_test)
# visualize wrongly scaled data
plt.scatter(X_train_scaled[:, 0], X_train_scaled[:, 1], c='b', label="training set", s=60)
plt.scatter(X_test_scaled_badly[:, 0], X_test_scaled_badly[:, 1], marker='^', c='r', label="test set", s=60)
plt.title("improperly scaled data")
plt.show()
```

Above images visualized data scaled improperly, as test and train are scaled separately. The test data points are moved inappropriately and the data looks a bit out of sync. 

Hence it is advised to scale both train and test data together on a similar scale. Another way to reason about this is the following: Imagine your test set was a single point. There is no way to scale a single point correctly, to fulfill the minimum and maximum requirements of the MinMaxScaler. But the size of your test set should not change your processing.
